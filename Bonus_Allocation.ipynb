```python
# Selecting only numerical columns
numerical_data = df.select_dtypes(include=['number'])

# Creating boxplots for each numerical column to identify outliers
plt.figure(figsize=(20, 15))

# Loop through each numerical column and create a subplot
for i, column in enumerate(numerical_data.columns, 1):
    plt.subplot(len(numerical_data.columns), 1, i)
    plt.boxplot(numerical_data[column], vert=False, patch_artist=True)
    plt.title(f'Boxplot for {column}')
    plt.xlabel(column)

plt.tight_layout()
plt.show()
```

```python
df.columns
```

```python
# List of columns to treat for outliers
columns_to_treat = [
    'Total_Amount_Wagered',
    'Average_Bet_Amount',
    'Number_of_Bonuses_Received',
    'Amount_of_Bonuses_Received',
    'Revenue_from_Bonuses',
    'Increase_in_Bets_After_Bonus',
    'Increase_in_wagering_after_Bonus'
]

# Apply capping to each column
for column in columns_to_treat:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Capping the outliers
    df[column] = np.where(df[column] < lower_bound, lower_bound,
                            np.where(df[column] > upper_bound, upper_bound, df[column]))
```

```python
# List of columns to check for outliers after treatment
columns_to_treat = [
    'Total_Amount_Wagered',
    'Average_Bet_Amount',
    'Number_of_Bonuses_Received',
    'Amount_of_Bonuses_Received',
    'Revenue_from_Bonuses',
    'Increase_in_Bets_After_Bonus',
    'Increase_in_wagering_after_Bonus'
]

# Creating boxplots for each numerical column to check if outliers are treated
plt.figure(figsize=(20, 15))

# Loop through each selected column and create a subplot
for i, column in enumerate(columns_to_treat, 1):
    plt.subplot(len(columns_to_treat), 1, i)
    plt.boxplot(df[column], vert=False, patch_artist=True)
    plt.title(f'Boxplot for {column}')
    plt.xlabel(column)

plt.tight_layout()
plt.show()
```

```python

from sklearn.preprocessing import MinMaxScaler

# Selecting only numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Initialize the scaler
scaler = MinMaxScaler()

# Fit and transform the numerical columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```
