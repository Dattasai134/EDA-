```python
# Get a list of numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns

# Print the numerical column names
numerical_cols
```

```python
def identify_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers

outliers_info = {}
for col in numerical_cols:
    outliers = identify_outliers_iqr(df[col])
    outliers_info[col] = {
        'count': len(outliers),
        'values': outliers.values.tolist()  # Convert to list for easier display
    }

for col, info in outliers_info.items():
    print(f"Outliers in column '{col}':")
    print(f"  Count: {info['count']}")
    if info['count'] > 0:
        print(f"  Values: {info['values']}")
    print("-" * 20)
```

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data

# Specify the columns to remove outliers from
columns_to_process = ['No. of Cards Printed', 'No. of Half Cards', 'No. of Quarter Cards', 'Accepted Cards', 'Rejected Cards']

# Apply the outlier removal function to each specified column
for column in columns_to_process:
    df = remove_outliers_iqr(df, column)

# Display the DataFrame after outlier removal (optional)
df
```
