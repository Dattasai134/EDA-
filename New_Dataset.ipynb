```python
import pandas as pd

df = pd.read_csv('COALINDIA 2019 - 2024.csv')
df
```

```python
df.duplicated().sum()
```

```python
import matplotlib.pyplot as plt

for col in df.columns:
  if col == 'Date':
    continue
  plt.figure()
  df[col].plot.box()
  plt.title(f'Boxplot for {col}')
  plt.show()
```

```python
# prompt: trim the outliers in every column

import numpy as np

for col in df.columns:
  if col == 'Date':
    continue
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  df[col] = np.clip(df[col], lower_bound, upper_bound)
```

```python
# prompt: now show me box plot for every column

import matplotlib.pyplot as plt
for col in df.columns:
  if col == 'Date':
    continue
  plt.figure()
  df[col].plot.box()
  plt.title(f'Boxplot for {col}')
  plt.show()
```
