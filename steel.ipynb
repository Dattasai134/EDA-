```python
from scipy import stats

# Calculate z-scores for all numeric columns excluding 'Sr_no' and 'Heat_no'
z_scores = stats.zscore(df[numeric_columns])

# Identify outliers (z-scores > 3 or < -3)
outliers = (z_scores > 3) | (z_scores < -3)

# Create a DataFrame to store outlier information
outlier_df = pd.DataFrame(outliers, columns=numeric_columns)

# Count the number of outliers in each column
outlier_counts = outlier_df.sum(axis=0)
print("Number of Outliers in Each Column:")
print(outlier_counts)

# You can also filter and display the rows with outliers
df_outliers = df[(outliers).any(axis=1)]
print("Rows with Outliers:")
print(df_outliers)
```

```python
# Calculate z-scores for numeric columns
z_scores = stats.zscore(df[numeric_columns])

# Identify outliers (z-scores > 3 or < -3)
outliers = (z_scores > 3) | (z_scores < -3)

# Create a boolean mask for rows without outliers
no_outliers_mask = ~(outliers).any(axis=1)

# Filter the DataFrame to remove outliers
df_cleaned = df[no_outliers_mask]

# Display the shape of the original and cleaned DataFrame
print(f"Original data shape: {df.shape}")
print(f"Cleaned data shape: {df_cleaned.shape}")

plt.figure(figsize=(20, 20))
for i, col in enumerate(numeric_columns):
    plt.subplot(len(numeric_columns) // 3 + 1, 3, i + 1)
    sns.boxplot(y=df_cleaned[col])
    plt.title(f'Boxplot of {col} (No Outliers)')
plt.tight_layout()
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
numeric_columns = numeric_columns.drop(['Sr_no', 'Heat_no'])

# Function to remove outliers using IQR
def remove_outliers_iqr(df, columns):
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1

    # Define outlier boundaries
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers
    df_cleaned = df[~((df[columns] < lower_bound) | (df[columns] > upper_bound)).any(axis=1)]
    return df_cleaned

# Remove outliers
df_cleaned = remove_outliers_iqr(df, numeric_columns)

# Display the shape of the original and cleaned DataFrame
print(f"Original data shape: {df.shape}")
print(f"Cleaned data shape: {df_cleaned.shape}")
```

```python
# Function to remove outliers using IQR
def remove_outliers_iqr(df, columns):
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1

    # Define outlier boundaries
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers
    df_cleaned = df[~((df[columns] < lower_bound) | (df[columns] > upper_bound)).any(axis=1)]
    return df_cleaned

# Apply the function iteratively until no more outliers are removed
df_cleaned = df.copy()
while True:
    df_before_cleaning = df_cleaned.copy()
    df_cleaned = remove_outliers_iqr(df_cleaned, numeric_columns)
    if df_cleaned.shape == df_before_cleaning.shape:
        break

# Display the shape of the original and final cleaned DataFrame
print(f"Original data shape: {df.shape}")
print(f"Final cleaned data shape: {df_cleaned.shape}")

# Visualize the cleaned data
plt.figure(figsize=(20, 20))
for i, col in enumerate(numeric_columns):
    plt.subplot(len(numeric_columns) // 3 + 1, 3, i + 1)
    sns.boxplot(y=df_cleaned[col])
    plt.title(f'Boxplot of {col} (No Outliers)')
plt.tight_layout()
plt.show()
```
