```python
# bonus_allocation_eda.py

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_excel("Bonus_Allocation.xlsx", sheet_name="Bonus Allocation Data - Master ")
```

```python
df
```

```python
# Basic Info
print("Dataset Shape:", df.shape)
print("\nDataset Info:\n", df.info())
print("\nMissing Values:\n", df.isnull().sum())
print("\nStatistical Summary:\n", df.describe())
```

```python
# Check duplicates
print("Duplicate rows:", df.duplicated().sum())
```

## **Exploratory Data Analysis**

```python
# Target variable distribution
sns.countplot(data=df, x="Should_Receive_Bonus")
plt.title("Target Variable Distribution")
plt.show()
```

```python
# Gender distribution
sns.countplot(data=df, x="gender", hue="Should_Receive_Bonus")
plt.title("Bonus Allocation by Gender")
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x="age", kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# Top 10 Countries with Most Users
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh', title="Top 10 Countries by User Count")
plt.xlabel("User Count")
plt.show()
```

# **Final Processed Dataset**

```python
# Export cleaned and processed data
df.to_csv("cleaned_bonus_allocation.csv", index=False)
print("Cleaned dataset saved as 'cleaned_bonus_allocation.csv'")
```
